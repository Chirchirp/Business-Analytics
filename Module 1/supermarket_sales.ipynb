```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("A:\Data World\Data\Data world\supermarket sales\supermarket_sales.csv")
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Days"] =df["Date"].dt.day_name()
df["Month"] = df["Date"].dt.month_name()
df['Weeks']=df['Days'].apply(lambda a: 'Weekday' if a=='Monday' or a=='Tuesday' or a=='Wednesday' or a=='Thrusday' or a=='Friday' else 'Weekend' )
df
```

```python
df.info()
```

```python
print(df.shape)
df['Gender'].value_counts()
```

```python
month_order = ['January', 'February', 'March']
```

```python
sns.countplot(x='Gender', data=df)
plt.show()

sns.countplot(x='Month', data=df)
plt.show()
```

```python

sns.countplot(x='Month', hue='Gender', data=df, order=month_order)
plt.title('Count of Gender per Month')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="Month", y="Total",hue='Gender', data=df, order=month_order, estimator=sum, ci=None)
for p in ax.patches:
    value = f'{p.get_height():,.2f}'  # Format value to 2 decimal places with comma separator
    ax.annotate(value, (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Total by Month')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday","Sunday"]
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="gross income", y="Days", data=df, order=day_order, estimator=sum, edgecolor='white', ci=None)
for p in ax.patches:
    value = f'{p.get_width():,.2f}'  # Format value to 2 decimal places with comma separator
    ax.annotate(value, (p.get_width(), p.get_y() + p.get_height() / 2.),
                ha='left', va='top', xytext=(10, 0), textcoords='offset points',
                color='Black')

plt.title('Gross income by per Day')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="Month", y="Total", data=df, order=month_order, estimator=sum, ci=None)

# Add formatted value annotations on the bars
for p in ax.patches:
    value = f'{p.get_height():,.2f}'  # Format value to 2 decimal places with comma separator
    ax.annotate(value, (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Total by Month')
plt.show()
```

```python
month_order = ['January', 'February', 'March']
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x="Month", y="Total", data=df, sort=month_order, estimator=sum)  

# Add formatted value annotations on the data points
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        value = f'{y:,.2f}'  # Format value to 2 decimal places with comma separator
        ax.annotate(value, (x, y),
                    ha='center', va='bottom', xytext=(0, 10), textcoords='offset points')

plt.title('Declining Sales for the 3 months')
plt.ylabel("Total Sales")
plt.show()
```

```python
month_order = ['January', 'February', 'March']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x="Month", y="gross income", hue="Product line", data=df, estimator=sum, ci=None)  

for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        value = f'{y:,.2f}'  # Format value to 2 decimal places with comma separator
        ax.annotate(value, (x, y),
                    ha='center', va='bottom', xytext=(0, 10), textcoords='offset points')
    
plt.title('Monthly Gross Income trends for Different Product Lines')
plt.xlabel('Month')
plt.ylabel('Gross Income')

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=3) 
ax.grid(True)
ax.grid(True, linestyle='--', alpha=0.7) 
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="Month", y="Total", hue= "City", data=df, order=month_order, estimator=sum, edgecolor='white', ci=None)
for p in ax.patches:
    value = f'{p.get_height():,.2f}'  # Format value to 2 decimal places with comma separator
    ax.annotate(value, (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Total sale per City')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
